```python
import pandas as pd
df = pd.read_csv("airline_passenger_satisfaction.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15,12))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
def label_encode(df, column_name, positive_class):
    df[column_name] = df[column_name].apply(lambda x: 1 if x == positive_class else 0)

label_encode(df, 'Gender', 'Male')
label_encode(df, 'Customer Type', 'Returning')
label_encode(df, 'Type of Travel', 'Business')
label_encode(df, 'Satisfaction', 'Satisfied')
```

```python
df = pd.get_dummies(df, columns=['Class'], prefix='Class')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15,12))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(6,4))
sns.barplot(data=df, x='Satisfaction', y='Class_Business')
plt.title("Satisfaction by Business Class")
plt.xticks([0, 1], ['Not Satisfied', 'Satisfied'])
plt.ylabel('% in Business Class')
plt.tight_layout()
plt.show()
```

```python
df['Age Group'] = pd.cut(df['Age'], bins=[0, 18, 35, 55, 100], labels=['<18','18-35','36-55','56+'])

plt.figure(figsize=(7,5))
sns.countplot(data=df, x='Age Group', hue='Satisfaction')
plt.title('Satisfaction by Age Group')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(data=df, x='Satisfaction', y='Flight Distance')
plt.title('Flight Distance vs Satisfaction')
plt.xticks([0, 1], ['Not Satisfied', 'Satisfied'])
plt.tight_layout()
plt.show()
```

```python
X = df.drop(['Satisfaction', 'ID', 'Age Group'], axis=1)
y = df['Satisfaction']
```
